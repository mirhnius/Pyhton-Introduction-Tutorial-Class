```python
import numpy as np
import matplotlib.pyplot as plt
# evenly sampled time at 200ms intervals
t = np.arange(0, 5, 0.2);

plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'r')
plt.plot(t,2*t+4,'g')

plt.show()
```

```python
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(10,5))

plt.subplot(131)
plt.bar(names, values)
plt.title('Plot1')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(132)
plt.scatter(names, values)
plt.title('Plot2')
plt.xlabel('a')
plt.ylabel('b')

plt.subplot(133)
plt.plot(names, values)
plt.title('Plot3')
plt.xlabel('i')
plt.ylabel('j')

plt.show()
```
